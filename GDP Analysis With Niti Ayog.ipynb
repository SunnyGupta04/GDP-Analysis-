# Insights from Correlation of GDP per capita with dropout rates in education (primary, upper primary and secondary) for the year 2014-2015Â¶

1. From the above plots we can tell that the drop rates is inversely proportional to the Per capita GSDP. So, if the drop rates    are high, per capita GSDP will be low and if drop rates are low, per capita GSDP will be high.

2. Drop out rates are higher in Secondary Education.

3. Drop out rates for Primary and Upper Primary are less than or equal to 15.

4. Drop out rates for Secondary is generally above 5.

5. Statistics for Primary and Upper Primary are pretty similar.

6. The outliers are due to Goa and Sikkim because of its Per Capita GDP. #### Hence, we can conclude that Drop out rates does      affect the per capita GSDP if it is not taken care of.
